# This JPN is about diff print() methods

```python
# basic print function
print("Hello, World!")
```

```python
# 2nd method of print function with f-string
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.") 
# ye jo place holder hain unhe use krke hum variable ki value print krte hain
```

```python
# 3rd method using string format() method
city = "New York"
country = "USA"
print("I live in {}, {}".format(city, country)) 
# almost like f-string but thoda different hai place holder khali rehta hai aur format() me variable pass krte hain
```

```python
# 4th method using percent % formatting
fruit = "apple"
quantity = 5
print("I have %d %s." % (quantity, fruit))
# ye purana method hai jisme %d for integers aur %s for strings use krte hain
```

```python
# 5th method using sys.stdout.write (new line end of string nahi deta hai)
import sys
sys.stdout.write("Hello world\n")
# isme hum manually \n use krke new line dete hain kyunki ye print() jaisa automatic new line nahi deta
```

```python
# 6th method using sep and end parameters
print("Hello", "World", sep=", ", end="!\n")
# sep parameter se hum alag karne wala character define krte hain 
# aur end parameter se string ke end me kya add karna hai wo define krte hain
```

```python
# 7th using triple quotes for multi-line strings
print("""This is line one.
This is line two.
This is line three.
""")
# isme f-string, format() ya % formatting use kar sakte hain multi-line strings ke liye bhi
```

```python
# 9th method printing objects with custom __str__ method
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self): # for printing vrna memory ki location print hoti hai
        return f"Person(name={self.name}, age={self.age})"

    def __repr__(self): # ye khud dekhnay k liye k kis object se data bana hai
        return f"Person(name={self.name!r}, age={self.age!r})"
p = Person("Bob", 25)
print(p)  # calls __str__
```
