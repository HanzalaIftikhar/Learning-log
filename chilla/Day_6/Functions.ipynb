```python
# user defined function for mean
def calculate_mean(numbers):
    """This function returns the mean of a list of numbers"""
    return round(sum(numbers) / len(numbers), 2)
```

```python
price = [200, 300, 400, 500, 600,433 , 343, 232, 123, 543, 654]
calculate_mean(price)
```

```python
# ex no 1
def sum_function(a, b):
    """This function returns the sum of two numbers"""
    return a + b

sum_function(int(input("Enter first number: ")), int(input("Enter second number: ")))
```

```python
# ex 2
def reverse_str_funcion(text):
    return text[::-1]

reverse_str_funcion(input("Enter name"))
```

```python
# ex no 3
def is_voter(age):
    return age >= 18


is_voter(int(input("Enter age")))

```

```python
# ex 4
def print_name(name, count=1):
    for i in range(count):
        print(name)


print_name("Hanzala", 3)
```

```python
# ex no 5

def cubid(len, width, height):
    return len*width*height

cubid(3,5,6)
```

```python
# ex no 6
def get_stats(list):
    mean = sum(list) / len(list)
    length = len(list)

    return mean, length

list = [23,43,54,65,76,87]
get_stats(list)
```

```python
# ex no 7

def print_even_numbers(n):
    for i in range(2, n + 1, 2):
        print(i)

print_even_numbers(10)
```
