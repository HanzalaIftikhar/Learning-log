## This JPN is about use input function


```python
age = input("Enter your age: ")
if age.isdigit():   
    age = int(age)
    if age >= 18:
        print("You are eligible to vote.")
    else:
        print("You are not eligible to vote.")
else:
    print("Invalid input. Please enter a numeric value for age.")

```

```python
# agar float ho input to hum is tarah kar sakte hain
age = input("Enter your age: ")
try:
    age = float(age)
    if age >= 18:
        print("You are eligible to vote.")
    else:
        print("You are not eligible to vote.")
except ValueError:
    print("Invalid input. Please enter a numeric value for age.")
```
