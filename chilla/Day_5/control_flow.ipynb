# This script is about control flow statements

```python
# conditonal operators in python
# >, <, >=, <=, ==, !=
# logical operators in python
# and, or, not
# control flow statements in python
# if, elif, else
```

```python
food = ['samosa', 'pizza', 'burger', 'pasta']
```

```python
# for loop to iterate through the list
for item in food:
    print(item)
```

```python
# while loop to print numbers from 0 to 5
i = 0
while i < 6:
    print(i)
    i += 1
```

```python
# control statement control with break
for item in food:
    if item == 'pasta': 
        break
    print(item)
```

```python
# control statement control with continue
for item in food:
    if item == 'burger': 
        continue
    print(item)
```

```python
# control flow statements with pass
for item in food:
    if item == 'pasta': 
        pass
    print(item)
```

```python
#  Nested if statement
input_str = input("Enter a number: ")
num = int(input_str)
if num > 0:
    if num % 2 == 0:
        print(f"{num} is a positive even number")
    elif num % 2 != 0:
        print(f"{num} is a positive odd number")
    else:
        print(f"{num} is a positive odd number")
```

# break and continue

```python
cup_of_chai = 0
while cup_of_chai < 5:
    cup_of_chai += 1
    print("I want more chai.")
    if cup_of_chai == 3:
        print("I am full now, no more chai.")
        break
    print("I have had", cup_of_chai, "cups of chai.")
```

## continue

```python
cup_of_chai = 0
while cup_of_chai < 5:
    cup_of_chai += 1
    print("I want more chai.")
    if cup_of_chai == 3:
        print("I am full now, no more chai.")
        continue
    print("I have had", cup_of_chai, "cups of chai.")
```

# infinite loop

```python
while True:
    print("This is an infinite loop.")
    break #be ware of infinite loops
```

```python
# anmpther way of handling infinite loop is by using a counter
count = 0
while True:
    print("This is loop number", count)
    count += 1
    if count == 5:
        break
```

## exception handling (try, except, finally)

```python
try:
    print("Hello")
    print(10 / 0)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
finally:
    print("This code will run anyway.")
```
